# 6. Price in Crore

```python
df['price_cr'].describe()
```

```python
sns.kdeplot(df['price_cr']);
```

```python
sns.boxplot(df['price_cr'],color='lightgreen');
```

```python

sns.histplot(df['price_cr'], bins=50, kde=True)

plt.title("Distribution of Price")
plt.xlabel("Price In Crore")
plt.ylabel("Frequency")
plt.show()
```

```python
Q1 = df['price_cr'].quantile(0.25)
Q3 = df['price_cr'].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
print("Skewness :", df['price_cr'].skew())
print("Kurtosis :", df['price_cr'].kurt())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
sns.histplot(df['price_cr'], bins=50, kde=True)
plt.title("Original Price Distribution")
plt.show()

plt.figure(figsize=(12,5))
sns.histplot(np.log(df['price_cr']), bins=50, kde=True, color='yellow')
plt.title("Log Transformed Price Distribution")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.ecdfplot(df['price_cr'],marker='.')

plt.title("ECDF of Property Prices")
plt.xlabel("Price (in Crores)")
plt.ylabel("Proportion")
plt.show()
```

```python
cv = df['price_cr'].std() / df['price_cr'].mean()
print("Coefficient of Variation:", cv)
```

## 📊 Price (in Crores) 

- Total number of properties: **2,668**
- Average price (Mean): **₹6.42 Cr**
- Median price: **₹3.95 Cr**
- Standard Deviation: **6.41**
- Minimum price: **₹0.40 Cr**
- Maximum price: **₹75 Cr**

### 🔎 Insights

- The **mean is higher than the median**, indicating the data is **positively skewed**.
- A few **high-value luxury properties** are increasing the average price.
- The middle 50% of properties lie between **₹2.49 Cr (Q1)** and **₹8.35 Cr (Q3)**.
- There is a **large variation in property prices**, suggesting the presence of outliers.
- The coefficient of variation is approximately 1, which means the standard deviation is almost equal to the mean.
  This indicates high relative variability in property prices. The market has a wide price range,
  likely due to the presence of both budget and luxury properties

Overall, the dataset shows a wide price distribution with some premium properties significantly affecting the overall average.
