# Sectors Vs Price 

```python
top_properties = df['property_name'].value_counts().head(7).index

df_top = df[df['property_name'].isin(top_properties)]

plt.figure(figsize=(8,5))
sns.barplot(data=df_top,
            x='property_name',
            y='price_cr',
            estimator=np.mean,palette='rocket')

plt.xticks(rotation=45)
plt.title("Median Price of Top 5 Most Listed Properties")
plt.show()
```

### The Price is House/Flat is higgest in Sector 109, 66 and 65.

```python
corr_matrix = df.corr(numeric_only=True)

price_corr = corr_matrix[['price_cr']].sort_values(by='price_cr', ascending=False)

plt.figure(figsize=(6,8))
sns.heatmap(price_corr,
            annot=True,
            cmap='coolwarm',
            center=0)

plt.title("Correlation with Price (price_cr)")
plt.show()
```
